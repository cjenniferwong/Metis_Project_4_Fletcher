```python
import glob
import pandas as pd
```

```python
list_of_df = []
for path in glob.glob('billboards_uri/*.pkl')[:]:
    df = pd.DataFrame.from_dict([pd.read_pickle(str(path))])
    list_of_df.append(df)
    #print(type(df))
    #break
    
uri_df = pd.concat(list_of_df)
```

```python
pd.to_pickle(uri_df, 'all_uri.pkl')
```
