```python

# gensim# gensi 
from gensim import corpora, models, similarities, matutils
# sklearn
from sklearn import datasets
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.cluster import KMeans
# logging for gensim (set to INFO)
import logging
logging.basicConfig(format='%(asctime)s : %(levelname)s : %(message)s', level=logging.INFO)
```

```python
text = pd.read_pickle('all_lyrics_with_org_names.pkl')
text_space = text['lyrics'].astype(str).values.tolist()
```

```python
from sklearn.feature_extraction import text

my_stop_words = text.ENGLISH_STOP_WORDS.union(["na", 'instrumental', 'yeah', 'ha', 'ok', 'babe', 'oo', 'baby', 
                                              'doo', 'ooh', 'da', 'hey', 'la', 'oh', 'whoa', 'choo', 'thoia', 
                                              'thoing', 'ah', 'nigga', 'niggas', 'boom', 'woo', 'chorus', 'wah',
                                              'eh', 'laa', 'll', 'ding', 'cha', 'ain', 'dee', 'don', 'shoo', 'ya',
                                              'yo', 'di', 'ba', 'hoo', 'ho', 'ha', 'cha', 've', 'fa', 'bwok',
                                              'bong', 'baby'])
```

```python
from nltk.stem.wordnet import WordNetLemmatizer
lmtzr = WordNetLemmatizer()

text_space

# Create a CountVectorizer for parsing/counting words# Creat 
count_vectorizer = CountVectorizer(ngram_range=(1, 2),  
                                   stop_words='english', token_pattern="\\b[a-z][a-z]+\\b")
count_vectorizer.fit(text_space)
```

```python
counts = count_vectorizer.transform(text_space).transpose()
counts.shape
```

```python
corpus = matutils.Sparse2Corpus(counts)
```

```python
id2word = { identifier: word for word, identifier in count_vectorizer.vocabulary_.items()}
len(id2word)
```

```python
lda = models.LdaModel(corpus=corpus, num_topics=5, minimum_probability=0.03, id2word=id2word, passes=10)
```
