```python
import pandas as pd
from glob import glob
```

```python
list_of_df = []

for path in glob('Lyrics_folder_818/*.pkl')[:]:
    df = pd.DataFrame.from_dict([pd.read_pickle(str(path))])
    list_of_df.append(df)

song_df = pd.concat(list_of_df)
```

```python
#pd.to_pickle(song_df, 'lyrics_df.pkl')

lyrics_df = pd.read_pickle('lyrics_df.pkl')
lyrics_df.head(10)
```

```python
names_df = pd.read_pickle('indexes_for_lyrics.pkl')
names_df.rename(columns = {
    
    'Song': 'song',
    'Artist': 'artist'
}, inplace= True)
```

```python
merged = pd.merge(lyrics_df, names_df, left_on = ['artist', 'song'], right_on = ['artist', 'song'], how = 'left')
```
