### Binning the genres

```python
import pandas as pd
```

```python
df = pd.DataFrame.from_dict(pd.read_pickle('more_spotify/first_5000.pkl'))
genre_list = df.genres.tolist()
genre_list[:5]
```

```python
slist =[]
for x in df.genres:
    try:
        #print(x)
        slist.extend(x)
    except:
        pass
```

```python
from collections import Counter
Counter(slist).most_common()
```

```python
chill = ['mellow gold', 'adult standards', 'quiet storm', 'urban contemporary','neo soul','lounge',
        'new wave','permanent wave',]
pop = ['dance pop', 'brill building pop' ,'pop rock', 'disco', 'pop', 'bubblegum pop', 'hip pop','pop rap',
      'new romantic','doo-wop','europop']
rnb = ['soul', 'r&b', 'motown', 'funk', 'classic soul','rap', 'hip hop','rhythm and blues','southern soul',
      ]
rock = ['rock', 'soft rock', 'album rock', 'folk rock', 'classic rock','new wave pop', 'rock-and-roll',
       'new jack swing', 'hard rock','dance rock','post-grunge',
       'roots rock','southern rock',]
christmas = ['christmas']
country = ['country', 'contemporary country', 'country road','rockabilly','folk','nashville sound','country rock',
          'heartland rock',]
```

```python
conversion_dict = {key: 'chill' for key in chill}
conversion_dict.update({key: 'pop' for key in pop})
conversion_dict.update({key: 'rnb' for key in rnb})
conversion_dict.update({key: 'rock' for key in rock})
conversion_dict.update({key: 'christmas' for key in christmas})
conversion_dict.update({key: 'country' for key in country})
conversion_dict
```

```python
from collections import Counter

genres = ['mell gold', 'do-wop', 'pp rap']
```

```python
%%time
[x[0] for x in Counter([conversion_dict[g] for g in genres if g in conversion_dict]).most_common(1)]
```

```python
def extract_genre_group(list_of_genres):
    return [x[0] for x in Counter([conversion_dict[g] for g in list_of_genres if g in conversion_dict]).most_common(1)]

df['genre_binned'] = df.genres.apply(lambda x: extract_genre_group(x) if type(x) is list else [])
len(df.genre_binned)
```
