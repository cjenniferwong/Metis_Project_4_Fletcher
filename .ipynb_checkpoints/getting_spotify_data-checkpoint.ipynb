```python
import requests

def get_tracks(user_id, playlist_uri, access_token):
    headers = {
        'Accept': 'application/json',
        'Content-Type': 'application/json',
        'Authorization': f'Bearer {access_token}',
    }

    # print(headers['Authorization'])
    r = requests.get(f'https://api.spotify.com/v1/users/{user_id}/playlists/{playlist_uri}/tracks?limit=100', headers=headers)

    return(r.json())
```
